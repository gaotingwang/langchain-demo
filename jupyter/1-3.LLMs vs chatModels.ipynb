```python
# 原生openai调用
from openai import OpenAI

client = OpenAI()
completion = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {
            "role": "user",
            "content": "你好，写首诗."
        }
    ]
)

print(completion.choices[0].message)
```

```python
#LLM调用，以openai为例
from langchain_openai import OpenAI
import os

os.environ["OPENAI_PROXY"] = "https://api.openai.com/v1"

api_base = os.getenv("OPENAI_PROXY")
api_key = os.getenv("OPENAI_API_KEY")

llm = OpenAI(
    model= "gpt-3.5-turbo",
    temperature=0,
    openai_api_key = api_key,
    openai_api_base = api_base
)

llm.invoke("你好")
```

```python
# 调用chatmodels，以openai为例
from langchain_openai import ChatOpenAI
from langchain.schema.messages import HumanMessage,AIMessage
import os

os.environ["OPENAI_PROXY"] = "https://api.openai.com/v1/chat"

api_base = os.getenv("OPENAI_PROXY")
api_key = os.getenv("OPENAI_API_KEY")

chat = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
    openai_api_key = api_key,
    openai_api_base = api_base
)

messages = [
    AIMessage(role="system",content="你好，我是tomie！"),
    HumanMessage(role="user",content="你好tomie，我是狗剩!"),
    AIMessage(role="system",content="认识你很高兴!"),
    HumanMessage(role="user",content="你知道我叫什么吗？")
]

response = chat.invoke(messages)
print(response)
```
